```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Distribution Plots

```python
sns.displot(tips['total_bill'], bins=10)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde')
```

```python
sns.pairplot(tips, hue='sex', palette='viridis')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

## Categorical Plots

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='coolwarm')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True, hue='sex', palette='rainbow')
```
