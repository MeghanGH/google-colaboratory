```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
x = plt.plot(x,y,'r--', color='green', linewidth=2)
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('My first graph')
```

```python
# subplot(nrows, ncols, index, **kwargs)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# add_axes(rect, projection=None, polar=False, **kwargs)
# left, bottom, width, height
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Lable')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0.1,0.5,0.5])
axes2 = fig.add_axes([0.2,0.2,0.3,0.15])

axes1.plot(y,x, 'g--')
axes1.set_title('Larger plot')

axes2.plot(x,y, 'r--')
axes2.set_title('Small plot')
```
